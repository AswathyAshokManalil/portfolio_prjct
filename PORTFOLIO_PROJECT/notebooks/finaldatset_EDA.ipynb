```python
import pandas as pd

file_path = r'C:/DS_mainproject/portfolio_prjct/PORTFOLIO_PROJECT/data/processed/final_dataset.csv'

# Step 1: Read the original file
final_df = pd.read_csv(file_path)

# Step 2: If Date is index, bring it back as column
final_df = final_df.reset_index()

# Step 3: Save properly (overwrite same file)
final_df.to_csv(file_path, index=False)

# Step 4: Read again properly
final_df = pd.read_csv(file_path, parse_dates=['Date'])

print(final_df.head(10))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Daily returns
sns.histplot(final_df['Daily_Return'], bins=50, kde=True)
plt.show()

# NIFTY return distribution
sns.histplot(final_df['NIFTY_Return'], bins=50, kde=True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Select only numeric columns for correlation
numeric_cols = final_df.select_dtypes(include=np.number)

# Plot correlation heatmap
plt.figure(figsize=(12,8))
sns.heatmap(numeric_cols.corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```
